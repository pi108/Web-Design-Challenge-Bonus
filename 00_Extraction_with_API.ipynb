```python
# Dependencies
import json
import requests 
from pprint import pprint
import pandas as pd

from api_keys import FRED_key

```

```python
# Save config information
url = "https://api.stlouisfed.org/fred/series/observations?series_id="

series_id = "DJIA"
api_key = FRED_key

# Build query URL
query_url = url + series_id + "&api_key=" + FRED_key + "&file_type=json" 

print(query_url)
```

```python
# Get data for specified economic indicator
response = requests.get(query_url)
response_json = response.json()
```

```python
# Verify that the response shows the number 200
print(response)
```

```python
# Print the response 
print(response_json)
```

```python
# Check the type of the response
print(type(response_json))
```

```python
# Display the keys for the response
print(response_json.keys())
```

```python
# Check the date of the response
print(response_json["realtime_start"])
```

```python
# Print the response in an easy-to-read format using pprint
pprint(response_json)
```

```python
# This is to identify that we have the correct JSON identified for the values we want to get

# The length we need to loop through is identified here
print("The total length that we have to loop through is")
print(len(response_json['observations']))
print (" ")

# Here we can see the values we are targeting
print("This is to test to make sure that we have the correct values")
print("values are " + response_json['observations'][0]['value'])
print("date is " + response_json['observations'][0]['date'])

# Now we need to start the loop
# Here we create the lists to store the data
date = []
value = []

length = len(response_json['observations'])

# Now let's loop!
for i in range(length):
    value.append(response_json['observations'][i]['value'])
    date.append(response_json['observations'][i]['date'])

# Now we create the dataframe
dji_df = {"Date":date,"DJIA":value}
dji_df = pd.DataFrame(dji_df)

# Here we print the dataframe header
print("The dataframe head is ")
print(" ")
print(dji_df.head())

# Here we confirm that the total number of rows in the dataframe matches the length of the response
print(" ")
print("Confirm the total number of rows matches the length")
print ("rows / columns ")
print(" ")
print(dji_df.shape)
```

```python
# Now we need to merge the multiple dataframes into one dataframe
# We will merge on date
# And we want to merge on an outer merge

combined_data = pd.merge(brent_df, dji_df, how='outer', on ='Date')
combined_data = pd.merge(combined_data,wti_df, how = 'outer', on ='Date')
combined_data = pd.merge(combined_data,hhng_df, how = 'outer', on ='Date')
combined_data = pd.merge(combined_data,gold_df, how = 'outer', on ='Date')
combined_data = pd.merge(combined_data,nasdaq_df, how = 'outer', on ='Date')
combined_data = pd.merge(combined_data,sp500_df, how = 'outer', on ='Date')
combined_data = pd.merge(combined_data,indfc_df, how = 'outer', on ='Date')
combined_data = pd.merge(combined_data,chi_to_usd_df, how = 'outer', on ='Date')
combined_data = pd.merge(combined_data,ind_to_usd_df, how = 'outer', on ='Date')
combined_data = pd.merge(combined_data,jp_to_usd_df, how = 'outer', on ='Date')
combined_data = pd.merge(combined_data,eu_to_usd_df, how = 'outer', on ='Date')

combined_data.head()

combined_data.to_excel("Output/00_Extraction/data_excel_original.xlsx", index = False)
combined_data.to_csv("Output/00_Extraction/data_csv_original.csv", index = False)
```
